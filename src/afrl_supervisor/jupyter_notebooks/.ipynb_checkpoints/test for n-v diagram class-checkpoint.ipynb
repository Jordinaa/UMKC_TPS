```python
import numpy as np
import matplotlib.pyplot as plt

A0 = 0.05984281113
CLA = 4.752798721
CLA_STALL = -3.85
CDA = 0.6417112299
CDA_STALL = -0.9233984055
ALPHA_STALL = 0.3391428111
WINGAREA = 0.6
AIR_DENSITY = 1.2041
MASS = 1.5
G = 9.81
```

```python
class FlightEnvelopeAssessment():
    def __init__(self, A0, CLA, CDA, ALPHA_STALL, WINGAREA, AIR_DENSITY, MASS, G, CLA_STALL, CDA_STALL):
        self.cla_stall = CLA_STALL
        self.cda_stall = CDA_STALL
        self.alpha0 = A0
        self.cla = CLA
        self.cda = CDA
        self.alpha_stall = ALPHA_STALL
        self.area = WINGAREA
        self.rho = AIR_DENSITY
        self.mass = MASS
        self.g = G

        self.coefficientLift = 0.0
        self.dynamicPressure = 0.0
        self.lift = 0.0
        self.velocity = 0.0
        self.loadFactor = 0.0
        self.stallSpeed = 0.0
        self.weight = 0.0

        self.vStall = 0.0
        self.clMaxWeights = [.2, .4, .6, .8]

        self.coefficientLiftList = []
        self.angleList = np.arange(0, ALPHA_STALL, 0.01 * np.pi/180)

        for angle in self.angleList:
            self.coefficientLift = self.calc_cl(angle)
            self.coefficientLiftList.append(self.coefficientLift)

        self.clMax = max(self.coefficientLiftList)
        self.angleListDegrees = np.rad2deg(self.angleList)
        self.calc_vStall()
        
# GRAPHS
    def calc_lift_coefficient_vs_aoa(self):
        plt.plot(self.angleListDegrees, self.coefficientLiftList)
        plt.xlabel("Angle of Attack (degrees)")
        plt.ylabel("Lift Coefficient (Cl)")
        plt.title("Lift Coefficient vs Angle of Attack")

        # Find the index of the ClMax in the coefficientLiftList
        cl_max_index = self.coefficientLiftList.index(self.clMax)

        # Find the corresponding angle of attack for ClMax
        alpha_cl_max = self.angleListDegrees[cl_max_index]

        # Plot the ClMax with a red dot
        plt.plot(alpha_cl_max, self.clMax, 'ro', label=f'ClMax: {self.clMax:.2f}')

        # Find the angle of attack for Vstall
        alpha_vstall = self.alpha_stall * 180 / np.pi

        # Find the Cl at Vstall
        cl_vstall = self.calc_cl(self.alpha_stall)

        # Plot the Cl at Vstall with a green dot
        plt.plot(alpha_vstall, cl_vstall, 'go', label=f'Cl at Vstall: {cl_vstall:.2f}')

        plt.legend(loc='best')
        plt.grid()
        plt.show()

        return self.clMax

    def calc_lift_vs_velocity(self):
        velocities = np.linspace(self.vStall, 1.5 * self.vStall, 100)
        lift_values = []

        for v in velocities:
            self.dynamicPressure = 0.5 * self.rho * v ** 2
            self.lift = self.clMax * self.area * self.dynamicPressure
            lift_values.append(self.lift)

        plt.plot(velocities, lift_values)
        plt.xlabel("Velocity (m/s)")
        plt.ylabel("Lift (N)")
        plt.title("Lift vs Velocity")
        plt.grid()
        plt.show()

    def calc_load_factor_vs_velocity(self):
        velocities = np.linspace(self.vStall, 1.5 * self.vStall, 100)
        load_factors = []

        for v in velocities:
            self.dynamicPressure = 0.5 * self.rho * v ** 2
            self.lift = self.clMax * self.area * self.dynamicPressure
            self.loadFactor = self.lift / (self.mass * self.g)
            load_factors.append(self.loadFactor)

        plt.plot(velocities, load_factors)

        # Calculate load factor at Vstall using clMax
        load_factor_vstall = self.clMax * 0.5 * self.rho * self.vStall ** 2 * self.area / (self.mass * self.g)

        # Plot the red dot at (Vstall, load_factor_vstall)
        plt.plot(self.vStall, load_factor_vstall, 'ro')

        plt.xlabel("Velocity (m/s)")
        plt.ylabel("Load Factor (n)")
        plt.title("Load Factor vs Velocity")
        plt.grid()
        plt.show()

    def calc_cl(self, angle_of_attack):
        cl = self.cla * (angle_of_attack - self.alpha0)
        return cl

    def calc_vStall(self):
        self.vStall = np.sqrt((2 * self.mass * self.g) / (self.rho * self.area * self.clMax))
```

```python
# Create a FlightEnvelopeAssessment object
fea = FlightEnvelopeAssessment(A0, CLA, CDA, ALPHA_STALL, WINGAREA, AIR_DENSITY, MASS, G, CLA_STALL, CDA_STALL)

# Calculate and plot Lift Coefficient vs Angle of Attack
cl_max = fea.calc_lift_coefficient_vs_aoa()
print("ClMax:", cl_max)

# Calculate and plot Lift vs Velocity
fea.calc_lift_vs_velocity()

# Calculate and plot Load Factor vs Velocity
fea.calc_load_factor_vs_velocity()
```

```python
class FlightEnvelopeAssessment():
    def __init__(self, A0, CLA, CDA, ALPHA_STALL, WINGAREA, AIR_DENSITY, MASS, G, CLA_STALL, CDA_STALL):
        self.cla_stall = CLA_STALL
        self.cda_stall = CDA_STALL
        self.alpha0 = A0
        self.cla = CLA
        self.cda = CDA
        self.alpha_stall = ALPHA_STALL
        self.area = WINGAREA
        self.rho = AIR_DENSITY
        self.mass = MASS
        self.g = G

        self.coefficientLift = 0.0
        self.dynamicPressure = 0.0
        self.lift = 0.0
        self.velocity = 0.0
        self.loadFactor = 0.0
        self.stallSpeed = 0.0
        self.weight = 0.0

        self.vStall = 0.0
        self.clMaxWeights = [.2, .4, .6, .8]

        self.coefficientLiftList = []
        self.angleList = np.arange(0, ALPHA_STALL, 0.01 * np.pi/180)

        for angle in self.angleList:
            self.coefficientLift = self.calc_cl(angle)
            self.coefficientLiftList.append(self.coefficientLift)

        self.clMax = max(self.coefficientLiftList)
        self.angleListDegrees = np.rad2deg(self.angleList)
        self.calc_vStall()

    def calc_cl(self, angle_of_attack):
        cl = self.cla * (angle_of_attack - self.alpha0)
        return cl

    def calc_vStall(self):
        self.vStall = np.sqrt((2 * self.mass * self.g) / (self.rho * self.area * self.clMax))

    def calc_lift_coefficient_vs_aoa(self):
        cl_max_index = np.argmax(self.coefficientLiftList)
        return self.angleListDegrees, self.coefficientLiftList, self.angleListDegrees[cl_max_index], self.coefficientLiftList[cl_max_index], self.clMax

    def calc_lift_vs_velocity(self):
        velocities = np.linspace(self.vStall, 1.5 * self.vStall, 100)
        lift_values = []

        for v in velocities:
            self.dynamicPressure = 0.5 * self.rho * v ** 2
            self.lift = self.clMax * self.area * self.dynamicPressure
            lift_values.append(self.lift)

        return velocities, lift_values

    def calc_load_factor_vs_velocity(self):
        velocities = np.linspace(self.vStall, 1.5 * self.vStall, 100)
        load_factors = []

        for v in velocities:
            self.dynamicPressure = 0.5 * self.rho * v ** 2
            self.lift = self.clMax * self.area * self.dynamicPressure
            self.loadFactor = self.lift / (self.mass * self.g)
            load_factors.append(self.loadFactor)

        return velocities, load_factors
```

```python
angleListDegrees, coefficientLiftList, aoa_cl_max, clMax = fea.calc_lift_coefficient_vs_aoa()
velocities, lift_values = fea.calc_lift_vs_velocity()
velocities, load_factors = fea.calc_load_factor_vs_velocity()
```

```python
# Lift Coefficient vs Angle of Attack
plt.plot(angleListDegrees, coefficientLiftList)
plt.xlabel("Angle of Attack (degrees)")
plt.ylabel("Lift Coefficient (Cl)")
plt.title("Lift Coefficient vs Angle of Attack")
plt.grid()
plt.show()

# Lift vs Velocity
plt.plot(velocities, lift_values)
plt.xlabel("Velocity (m/s)")
plt.ylabel("Lift (N)")
plt.title("Lift vs Velocity")
plt.grid()
plt.show()

# Load Factor vs Velocity
plt.plot(velocities, load_factors)
plt.xlabel("Velocity (m/s)")
plt.ylabel("Load Factor (n)")
plt.title("Load Factor vs Velocity")
plt.grid()
plt.show()
```
