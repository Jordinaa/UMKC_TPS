```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Define the time series data
y = np.random.randn(1000)

# Define the time values
t = np.linspace(0, 1, 1000)

# Define the filter order and cutoff frequency
order = 4
cutoff = 0.1
fs = 1 / np.mean(np.diff(t))

# Create the Butterworth filter coefficients
b, a = signal.butter(order, cutoff / (fs / 2), 'low')

# Apply the filter to the time series data
y_filtered = signal.filtfilt(b, a, y)

# Plot the original and filtered signals
plt.figure()
plt.plot(t, y, label='Original Signal')
plt.plot(t, y_filtered, label='Filtered Signal')
plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('Signal Amplitude')
plt.show()
```
