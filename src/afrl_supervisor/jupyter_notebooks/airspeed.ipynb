```python
import numpy as np
import matplotlib.pyplot as plt

# Constants and variables
A0 = 0.05984281113
CLA = 4.752798721
CLA_STALL = -3.85
CDA = 0.6417112299
CDA_STALL = -0.9233984055
ALPHA_STALL = 0.3391428111
WINGAREA = 0.6
AIR_DENSITY = 1.2041
MASS = 1.5
G = 9.81

# Calculate lift coefficient (CL) at the stall condition
CL_STALL = CLA * (ALPHA_STALL - A0)
print(CL_STALL)
# Calculate velocity (V) for a range of load factors (n)
load_factors = np.linspace(1, 5, 100)  # Change the range and number of points as needed
velocities = []

for n in load_factors:
    # Lift force (L) equals the load factor (n) times the aircraft weight (W)
    L = n * MASS * G

    # Calculate the dynamic pressure (q) using the lift equation: L = 0.5 * AIR_DENSITY * V^2 * WINGAREA * CL_STALL
    q = L / (0.5 * WINGAREA * CL_STALL)

    # Calculate the velocity (V) from the dynamic pressure: V = sqrt(2 * q / AIR_DENSITY)
    V = np.sqrt(2 * q / AIR_DENSITY)
    velocities.append(V)

# Plot the load-factor vs. velocity graph
plt.plot(velocities, load_factors)
plt.xlabel("Velocity (m/s)")
plt.ylabel("Load Factor (n)")
plt.title("Load-Factor vs. Velocity")
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Constants and variables
A0 = 0.05984281113
CLA = 4.752798721
CLA_STALL = -3.85
CDA = 0.6417112299
CDA_STALL = -0.9233984055
ALPHA_STALL = 0.3391428111
WINGAREA = 0.6
AIR_DENSITY = 1.2041
MASS = 1.5
G = 9.81

# Calculate lift coefficient (CL) at the stall condition
CL_STALL = CLA * (ALPHA_STALL - A0)

# Calculate the aircraft's weight (W)
W = MASS * G

# Calculate stall speed (V_S)
V_S = np.sqrt((2 * W) / (AIR_DENSITY * WINGAREA * CL_STALL))

print("Stall speed (V_S):", V_S, "m/s")
```
