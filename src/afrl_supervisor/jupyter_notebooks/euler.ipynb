```python
import numpy as np
import matplotlib.pyplot as plt

A0 = 0.05984281113 
CLA = 4.752798721
CLA_STALL = -3.85 
CDA = 0.6417112299
CDA_STALL = -0.9233984055
ALPHA_STALL = 0.3391428111
alpha_stall_deg = np.rad2deg(ALPHA_STALL)
WINGAREA = 0.6
AIR_DENSITY = 1.2041
MASS = 1.5
G = 9.81
```

```python
calc_cl = lambda angle: CLA * (np.deg2rad(angle)-A0)

calc_dyn_press = lambda v: .5 * AIR_DENSITY * (v**2)

calc_lift = lambda angle, v: calc_cl(angle) * calc_dyn_press(v) * WINGAREA 

calc_load_factor = lambda angle, v: (calc_lift(angle, v) * calc_dyn_press(v) * WINGAREA)/(MASS * G)
```

```python
cl_list = []
lift_list = []
dyn_list = []
velocity_range = np.arange(50)
velocity_list = []
angle_list = np.linspace(0, alpha_stall_deg, 50)
load_factor_list = []

velocity = np.linspace(0, 100, 250)

# for angle in angle_list:
#     CL = calc_cl(CLA, angle, A0)
#     cl_list.append(CL)
#     ang_deg_list.append(angle)

angle1 = 0
    
for vel in velocity_range:
    n = calc_load_factor(angle1, vel)
    load_factor_list.append(n)
    velocity_list.append(vel)
    
# clMax = max(cl_list)
# print(f"clmax: {clMax}")


print(load_factor_list)
plt.plot(velocity_list, load_factor_list, label = 'lift curve slope')
plt.xlabel('AoA')
# plt.ylim(-.5, 1.5)
plt.ylabel('Cl')
plt.title('Cl vs. AoA')
plt.legend(loc='best')
plt.grid(True)
plt.show()    
```
